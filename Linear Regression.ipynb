```python
df.drop(["acceleration", "displacement"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
from datetime import datetime
today = datetime.today()
today
```

```python
this_year = today.year
this_year
```

```python
df["age"] = this_year - df.model_year
```

```python
df.head()
```

```python
df["origin_japan"] = df["origin_japan"].astype(int)
```

```python
df["origin_usa"] = df["origin_usa"].astype(int)
```

```python
df.head()
```

**Model**

```python
x = df.drop(["mpg"], axis = 1)
y = df["mpg"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm_model = LinearRegression()
```

```python
lm_model.fit(x_train, y_train)
```

```python
lm_model.intercept_
```

```python
y_pred = lm_model.predict(x_test)
```

```python
y_pred.shape
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
r2 = r2_score(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
MAE = mean_absolute_error(y_test, y_pred)
intercept = lm_model.intercept_
r2, MSE, MAE, intercept
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_test, y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color = "red")
plt.show()
```

```python
new_data = np.array([[8, 150, 3500, 55, 1, 0]])  # example values
predicted_mpg = lm_model.predict(new_data)
print("Predicted MPG:", predicted_mpg)
```
